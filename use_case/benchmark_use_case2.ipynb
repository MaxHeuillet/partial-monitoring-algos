```python
import experiment

n_labels = 10
horizon  = 10000
alpha = 1.01 #cbp hyperparameter
proba = 0.01 #probability of the confidence intervals
threshold = 0.025 #alarm threshold


### Wald confidence interval:
z_value = norm.ppf(1 - proba/2)
margin = threshold / 10
prior = 0.1
nsample_wald = z_value**2 * ( prior * (1-prior) ) / margin**2 
print('Wald threshold', nsample_wald)

game = games.label_efficient2( threshold )

stream = 'imbalanced'
matrix = 'nonuniform'

task = experiment.Evaluation( n_labels,stream,matrix,horizon) 

alpha = 1.01 #cbp hyperparameter
nsample_cbp = np.floor( np.exp( 1/(1-2*alpha) * np.log(proba/4) ) ) + 1
print('nsample cbp', nsample_cbp)
alg = utils.CBPstrategy(game, False, n_labels, alpha, nsample_cbp)
# alg = utils.KFirst2(game, n_labels, nsample_wald, threshold)


seed = 89
np.random.seed(seed)
imbalance, M, ground_truth = utils.get_config(matrix,stream, n_labels)
labels, preds = experiment.Simulation(M, imbalance ).get_contexts(horizon,)
res = task.eval_policy_once(game, alg, seed)

print([ 0 if i>threshold else 1 for i in ground_truth  ], res[4])
```
