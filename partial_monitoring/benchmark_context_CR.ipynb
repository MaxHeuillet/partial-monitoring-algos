```python
import pandas as pd
import numpy as np
from scipy import stats
from scipy.stats import ttest_ind,mannwhitneyu

dfs = []
n_folds = 96
horizon = 20000


def loadall_results(path, horizon, n_folds):
    result = np.zeros( (n_folds, horizon) )
    with gzip.open(  path ,'rb') as f:
        for i in range(n_folds):
            try:
                result[i] = pkl.load(f)[:horizon]
            except EOFError:
                break
    return result

for game in ['AT', 'LE']:

    for task in [ 'balanced']:

        if game == 'AT':
            labels = [ 'RandCBPside2', 'CBPside','PGTS', 'PGIDSratio', 'STAP', 'random'  ]   
        else:
            labels = [ 'RandCBPside2', 'CBPside', 'CesaBianchi', 'random'  ] 

        perf_dic = {}

        direct = './contextual_results/{}'.format(game)

        for label in  labels:


            if game == 'AT':

                if label == 'RandCBPside2':
                    path = os.path.join(direct, 'cr_benchmark_{}_{}_20000_96_RandCBPside2_1_5_07.pkl.gz'.format(task,context_type) )
                    result = loadall_results(path, 19999, 96)

                elif label == 'CBPside':
                    path = os.path.join(direct, 'cr_benchmark_{}_{}_20000_96_CBPside.pkl.gz'.format(task,context_type) )
                    result = loadall_results(path, 19999, 96)

                elif label == 'random':
                    path = os.path.join(direct, 'cr_benchmark_balanced_linear_20000_96_random.pkl.gz'.format(task,context_type) )
                    result = loadall_results(path, 19999, 96)

                elif label == 'PGIDSratio':
                    path = os.path.join(direct, 'cr_benchmark_balanced_linear_7500_96_PGIDSratio.pkl.gz'.format(task,context_type) )
                    result = loadall_results(path, 7499, 96)
                
                elif label == 'PGTS':
                    print(game)
                    path = os.path.join(direct, 'cr_benchmark_balanced_linear_7500_96_PGTS.pkl.gz'.format(task,context_type) )
                    result = loadall_results(path, 7499, 96)

                elif label == 'STAP':
                    path = os.path.join(direct, 'cr_benchmark_balanced_linear_20000_96_STAP.pkl.gz'.format(task,context_type) )
                    result = loadall_results(path, 19999, 96)


            else:

                if label == 'RandCBPside2':
                    path = os.path.join(direct, 'cr_benchmark_{}_{}_20000_96_RandCBPside2_1_5_07.pkl.gz'.format(task,context_type) )
                    result = loadall_results(path, 19999, 96)

                elif label == 'CBPside':
                    path = os.path.join(direct, 'cr_benchmark_{}_{}_20000_96_CBPside.pkl.gz'.format(task,context_type) )
                    result = loadall_results(path, 19999, 96)
                    
                elif label =='CesaBianchi': 
                    path = os.path.join(direct, 'cr_benchmark_balanced_linear_20000_96_CesaBianchi.pkl.gz'.format(task,context_type) )
                    result = loadall_results(path, 19999, 96)

                elif label == 'random':
                    path = os.path.join(direct, 'cr_benchmark_balanced_linear_20000_96_random.pkl.gz'.format(task,context_type) )
                    result = loadall_results(path, 19999, 96)


            print(result.shape)
            perf_dic[label] = result[:,-1] 

        df = pd.DataFrame.from_dict(perf_dic)

        ############### create data frame:

        # Calculate mean and std
        mean_values = df.mean()
        std_values = df.std()

        # Calculate p-values
        p_values = {}
        for strategy in df.columns:
            if strategy == 'RandCBPside2':
                p =  1
                p_values[strategy] = p
            else:
                s, p = ttest_ind(df['RandCBPside2'], df[strategy], equal_var=False)
                # s, p = mannwhitneyu(df['RandCBP'], df[strategy],  alternative='less')
                # s, p = ttest_ind( df[strategy], df['RandCBP'])
                p_values[strategy] = p

        # Find the minimum value within each row
        min_values = df.min(axis=1)
        # Create a boolean DataFrame indicating if each strategy achieved the minimum value
        strategy_won = df.eq(min_values, axis=0)
        # Count how many times each strategy won
        winning_counts = strategy_won.sum()
        # winning_counts = df.apply(idxmin_excluding_duplicates, axis=1).value_counts()

        # Create the new DataFrame
        new_data = {
            'Mean_{}_{}'.format(game,task): mean_values,
            'Std_{}_{}'.format(game,task): std_values,
            'PValue_{}_{}'.format(game,task): p_values,
            'WinningCount_{}_{}'.format(game,task): winning_counts,  }

        pd.options.display.float_format = '{:.3f}'.format
        
        new_df = pd.DataFrame(new_data)
        new_df['WinningCount_{}_{}'.format(game,task)].fillna(0, inplace=True)
        new_df['WinningCount_{}_{}'.format(game,task)] = new_df['WinningCount_{}_{}'.format(game,task)].astype(int) 
        dfs.append(new_df)
```

```python
df = pd.concat(dfs, axis=1)
# df = df.round(2)

# Sample DataFrame columns
columns = [
    'Mean_AT_balanced', 'Std_AT_balanced', 'Winning Count_AT_balanced', 'P-Value_AT_balanced',
    'Mean_LE_balanced', 'Std_LE_balanced', 'Winning Count_LE_balanced', 'P-Value_LE_balanced'
]

# Create arrays for each level of the multi-index
first_level = ['mean', 'std', 'pvalue', 'win count',] * 2
second_level = [ 'balanced', 'balanced', 'balanced', 'balanced'] * 2
third_level = ['AT', 'AT', 'AT', 'AT', 'LE', 'LE', 'LE', 'LE'] 

# Create a MultiIndex from the arrays
multi_index = pd.MultiIndex.from_arrays([third_level , second_level, first_level], names=('Game', 'Case', 'Metric'))

# # Create a DataFrame with the specified columns and multi-index
# # df = pd.DataFrame(columns=columns, index=[0])
df.columns = multi_index

desired_order = [ 'RandCBPside2', 'CBPside',  'PGIDSratio', 'PGTS', 'STAP', 'CesaBianchi', 'random' ] # 'PGIDSratio', 'PGTS',
df = df.reindex(desired_order)

df.index = [ 'RandCBPside', 'CBPside', 'PGIDSratio', 'PGTS', 'STAP', 'CESA', 'random'  ] #'PGIDSratio', 'PGTS',

df = df.round(3)

df
```
